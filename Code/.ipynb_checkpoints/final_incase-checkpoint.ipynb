## Part 3: Features engineering:

### Transforming categorical ordinal variables into numerical ordinal variables:

Some variables are ordinate with special text values like good or bad, we replace them that are encoded this way with number with 5 for the higher score, 1 for lowest and 0 where this value is missing.

```python
CatColsOrd = ('ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'HeatingQC', 'KitchenQual',
             'FireplaceQu', 'GarageQual', 'GarageCond', 'PoolQC')
```

```python
def EncodingCatVar(x,c):
    for cols in c:
        x[cols] = x[cols].str.upper()
        x.loc[x[cols] == "EX", cols] = 5
        x.loc[x[cols] == "GD", cols] = 4
        x.loc[x[cols] == "TA", cols] = 3
        x.loc[x[cols] == "FA", cols] = 2
        x.loc[x[cols] == "PO", cols] = 1
        x.loc[x[cols] == "NO", cols] = 0

```

```python
EncodingCatVar(train, CatColsOrd)
EncodingCatVar(test, CatColsOrd)
```

For the others that might be encoded specificly, we will use the dummy approach instead.

### Transforming numerical variable that are in fact categorical:

We identified the three following variables as coded as numerical but they don't have any order of the values. It's means that in fact they are really categorical variable that were encoded with number. In our model, it can lead to giving more importance to an higher value (for example december = 12) that will make december more important than January (=1) where there are equal. To fix this, we will transform them into string:
* MSSubClass: 
* MoSold: 
* YrSold:

```python
train['MoSold'] = train['MoSold'].astype(str)
train['YrSold'] = train['YrSold'].astype(str)
train['MSSubClass'] = train['MSSubClass'].astype(str)

test['MoSold'] = test['MoSold'].astype(str)
test['YrSold'] = test['YrSold'].astype(str)
test['MSSubClass'] = test['MSSubClass'].astype(str)
```

So now, we will be able to dummy encode them later.

### Creation of new columns:

We can create the total house square floor area by summing the total basement area + the total 1st floor + the total second floor:

```python
train['TotalSquareFloor'] = train['TotalBsmtSF'] + train['1stFlrSF'] + train['2ndFlrSF']

test['TotalSquareFloor'] = test['TotalBsmtSF'] + test['1stFlrSF'] + test['2ndFlrSF']
```

We can also compute the average score of all the categorical ordinal variables that we transformed before:

```python
train['AvgScoreCatOrd'] = train[['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'HeatingQC', 'KitchenQual',
             'FireplaceQu', 'GarageQual', 'GarageCond', 'PoolQC']].mean(axis = 1)

test['AvgScoreCatOrd'] = test[['ExterQual', 'ExterCond', 'BsmtQual', 'BsmtCond', 'HeatingQC', 'KitchenQual',
             'FireplaceQu', 'GarageQual', 'GarageCond', 'PoolQC']].mean(axis = 1)
```

We compute the total number of bathrooms by adding the bathrooms variables and 0.5 to the ones which correspond to half bathrooms:

```python
train['Total_Bathrooms'] = (train['FullBath'] + (0.5 * train['HalfBath']) +
                               train['BsmtFullBath'] + (0.5 * train['BsmtHalfBath']))

test['Total_Bathrooms'] = (test['FullBath'] + (0.5 * test['HalfBath']) +
                               test['BsmtFullBath'] + (0.5 * test['BsmtHalfBath']))
```

We compute also the total porch square feet by summing all variables related:

```python
train['Total_porch_sf'] = (train['OpenPorchSF'] + train['3SsnPorch'] +
                              train['EnclosedPorch'] + train['ScreenPorch'] +
                              train['WoodDeckSF'])

test['Total_porch_sf'] = (test['OpenPorchSF'] + test['3SsnPorch'] +
                              test['EnclosedPorch'] + test['ScreenPorch'] +
                              test['WoodDeckSF'])
```

And finally the total square feet as the sum of GrLivArea and ToTalBsmtSF:

```python
train['TotalSquareFeet'] = train['GrLivArea'] + train['TotalBsmtSF']

test['TotalSquareFeet'] = test['GrLivArea'] + test['TotalBsmtSF']
```

### Manage Skewness for predictors:

Now we're going to compute skewness for all numeric features in order to see if we need to apply a log transformation:

We first select all the numerical observations:

```python
numeric_feats = train.select_dtypes(include=[np.number])
```

```python
numeric_feats.shape
```

We have 50 numeric columns:

We now compute the Skewness and order the score:

```python
NumSkew = numeric_feats.skew(axis = 0, skipna = True, numeric_only = True) 
NumSkew = NumSkew.sort_values(ascending = False)
```

And assign it into a dataframe:

```python
NumSkew = pd.DataFrame(NumSkew)
NumSkew = NumSkew.reset_index()
NumSkew.columns = ["Features", "SkewScore"]
```

And print it:

```python
NumSkew.head(50)
```

Now, for all features with a skewscore above 0.5 or bellow -0.5, we're going to apply a log transformation:

```python
SkewToFix = NumSkew.loc[abs(NumSkew['SkewScore'])>0.5,]
```

```python
SkewToFix.shape
```

We have now 38 features where we will apply a log transformation:

```python
for i in SkewToFix['Features']:
    train[i] = np.log1p(train[i])
    test[i] = np.log1p(test[i])
```

And we check again the skewness after logged transformed it:

```python
numeric_featsLoged = train.select_dtypes(include=[np.number])
NumSkewLoged = numeric_featsLoged.skew(axis = 0, skipna = True, numeric_only = True) 
NumSkewLoged = NumSkewLoged.sort_values(ascending = False)
NumSkewLoged = pd.DataFrame(NumSkewLoged)
NumSkewLoged = NumSkewLoged.reset_index()
NumSkewLoged.columns = ["Features", "SkewScore"]
print("Skewness average before log transformation:",NumSkew.mean())
print("Skewness average after log transformation:",NumSkewLoged.mean())
```

We can see that even if skewness remains, we decrease it so our variable transformation lead to less variability in the data which will increase the performance of our models.

## Compare the score based on RMSE:

We could use the cross_val_score function from sklearn to compute the score of our models with 10 folds but here with the bagging model, it's kind of hard as it's not a model but just an average.

So we need to create a function to compare models and bagging:

We first create a function to return the Root mean squared error:

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
import statistics


def RMSE(Pred,Real):
    RMSE = sqrt(mean_squared_error(Pred, Real))
    return(RMSE)
```

Now we initialize a list of models we want to try and also empty list that we will use in our homemade functions to compare the score of our models and bagging.

```python
models = {lasso:"lasso",
          LGBModel : "LGBM",
          ENet : "Elastic Net",
          xgb_model : "XGBoost"
         }
rmse = []
fold = 10
PredList = []
PredBaggedRMSE=[]
```

We create a function that return the average root mean squared error for all the models in the dictionnary models. Also, it will get the predictions of all model in a list and averaging them each fold in the cross validation and will return the root mean squared error. At the end, we compute the average score of the root mean squared error for the bagging and it will lead to be able to compare performance of our models.

Note: we use the random_state parameters in the train_test_split function in order to make sure that it's the same split which is done for the models and bagging part:

### Note: As above, we do not need to transform the data to get the log root mean squared error as our target variable is already logged transformed.

```python
def ModelsEvaluation(models, data):
        for i in models:
            for x in range(0,fold):
                trainX , testX = train_test_split(data,test_size=0.2, random_state = x)

                i.fit(X=trainX[selected], y=trainX[target])
                pred = i.predict(testX[selected])
                rmse.append(RMSE(pred, testX[target]))


            ScoreRMSE = np.mean(rmse)
            print(ScoreRMSE,"is the cross validated score with",fold,"folds for the",models[i],"model")
    
        for x in range(0,fold):
            trainX , testX = train_test_split(data,test_size=0.2, random_state = x)

            for i in models:
                i.fit(X=trainX[selected], y=trainX[target])              
                PredList.append(i.predict(testX[selected]))

            PredBagged = np.mean(PredList, axis = 0)
            PredBaggedRMSE.append(RMSE(PredBagged, testX[target]))
            PredList.clear()
        
        BaggingScore = np.mean(PredBaggedRMSE, axis = 0)
        print(BaggingScore,"is the cross validated score of bagging", str(models.values()))

```

And we compute the score:

```python
random.seed(1)
ModelsEvaluation(models, train)
```
